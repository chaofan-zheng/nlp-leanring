## matplotlib 模块的使用

先要安装第三方库，在终端中使用命令```pip install --user matplotlib```

- 普通折线图 plot

````python
from matplotlib import pyplot as plt  # 里面都是封装好的函数
````

````python
x_list = [1,2,3,4,5,6]
y_list = [1,4,9,16,25,36]
plt.plot(x_list,y_list)  # x,y 
plt.show()
````

- 散点图 scatter

````python
x_tuple,y_tuple = (1,2,3,4,5,6),(1,4,9,16,25,36)
plt.scatter(x_tuple,y_tuple) # scatter
plt.show()
````

## 尝试使用更大规模的数据

````python
x_values = list(range(1,1001))
y_values = [x**2 for x in x_values]
plt.scatter(x_values,y_values)
plt.show()
````

## 自定义表标题，坐标等

````python
plt.plot(x_values,y_values,c = 'red') # 颜色也可以用RGB格式设定
plt.title("squre number",fontsize = 14) # plt.title("")
plt.xlabel("x values")  xlabel
plt.ylabel("y values") # 不支持中文
plt.tick_params(axis="both",labelsize = 8) # "both","x","y"
plt.show() 
````

## 叠加图

````python
x_values = list(range(1,10))
y_values = [x**2 for x in x_values]
y02_values = [x**3 for x in x_values]
plt.plot(x_values,x_values,label="y = x")
plt.plot(x_values,y_values,label="y = x**2")
plt.plot(x_values,y02_values,label= "y = x**3")
plt.legend(loc = "best")  #调整图例位置
plt.show()
````

## 保存图表

````python
plt.plot(x_values,y_values,c = 'red') # 颜色也可以用RGB格式设定
plt.title("squre number",fontsize = 14)
plt.xlabel("x values")
plt.ylabel("y values") # 不支持中文
plt.tick_params(axis="both",labelsize = 8) # "both","x","y"
# 保存
plt.savefig("test.jpeg")
````

## 柱状图

````python
x_values = list(range(1,10))
y_values = [x**2 for x in x_values]
plt.bar(x_values,y_values,)
plt.show()
````

## 饼状图

````python
name = ["A","B","C","D"]
values = [11,23,55,2]
plt.pie(values,labels = name,autopct='%.1f %%') # 格式：保留一位小数，空格，输出百分号
plt.legend(loc = "best") # legend(loc= "best")
plt.show()
````
